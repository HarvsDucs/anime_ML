```python
# Checking the categorical columns
idx1 = df.columns
idx2 = df.describe().columns

def compare_indexes(idx1, idx2):
    # Find columns unique to first index
    unique_to_idx1 = idx1.difference(idx2)
    
    # Find columns unique to second index
    unique_to_idx2 = idx2.difference(idx1)
    
    # Find common columns
    common = idx1.intersection(idx2)
    
    return {
        'unique_to_first': sorted(unique_to_idx1),
        'unique_to_second': sorted(unique_to_idx2),
        'common': sorted(common),
        'summary': {
            'total_in_first': len(idx1),
            'total_in_second': len(idx2),
            'common_count': len(common),
            'unique_to_first_count': len(unique_to_idx1),
            'unique_to_second_count': len(unique_to_idx2)
        }
    }

compare_indexes(idx1, idx2)
```
